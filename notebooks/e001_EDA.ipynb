```python
from sklearn.model_selection import GroupShuffleSplit
```

```python
gsf = GroupShuffleSplit(n_splits=5, random_state=71)
gsf.split(trn_meta_df.drop('target', axis=1), trn_meta_df.target, groups=trn_meta_df.id_measurement)
```

```python
for trn_idx, val_idx in gsf.split(trn_meta_df.drop('target', axis=1), trn_meta_df.target, groups=trn_meta_df.id_measurement):
    print(trn_idx)
```

```python
[fold for fold in gsf.split(trn_meta_df.drop('target', axis=1), trn_meta_df.target, groups=trn_meta_df.id_measurement)]
```

```python
from itertools import tee

folds = gsf.split(trn_meta_df.drop('target', axis=1), trn_meta_df.target, groups=trn_meta_df.id_measurement)
folds, _folds = tee(folds)
for i, j in folds:
    print(i)
    
print('==================================')    
for i, j in _folds:
    print(i)
    print(len(i))
    print(len(j))
    print(np.sum(trn_meta_df.target.values[j]))
print('==================================')    
```
